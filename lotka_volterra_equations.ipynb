```python
# Frameworks
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%time
# Lotka-Volterra Equations

# Time
start_time = 0
end_time = 50
dt = 0.001
time = np.arange(start_time, end_time, dt)

# Initial Population
x, y = 1, 0.5

# Parameters Equations
alpha, beta = 0.7, 0.5 # prey
delta, gamma = 0.3, 0.2 # predator

# Populations
prey_pop, predator_pop = [x], [y]
for count in range(len(time)-1):

    x = x + ( (alpha * x) - (beta * x * y) ) * dt
    y = y + ( (delta * x * y) - (gamma * y) ) * dt

    prey_pop.append(x)
    predator_pop.append(y)
```

```python
# Graphs: Lotka-Volterra
plt.plot(time, prey_pop, color="green", linestyle="--", label="prey")
plt.plot(time, predator_pop, color="red", linestyle="--", label="predator")
plt.title("Lotka-Volterra")
plt.xlabel("x")
plt.ylabel("y")
plt.xlabel("time")
plt.ylabel("population")
plt.legend(loc="upper left")
plt.grid(linestyle="--", alpha=0.5, color="gray")
```
